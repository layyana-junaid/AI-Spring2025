```python
# Tree Representation
tree = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G'],
    'D': ['H', 'I'],
    'E': ['J'],
    'F': ['K', 'L'],
    'G': [],
    'H': [],
    'I': ['M', 'N'],
    'J': [],
    'K': [],
    'L': [],
    'M': [],
    'N': [],
}

# BFS Function
def bfs(graph, start, goal):
    visited = []  # List for visited nodes
    queue = []  # Initialize a queue
    visited.append(start)
    queue.append(start)

    while queue:
        node = queue.pop(0)  # Dequeue
        print(node, end=" ")
        
        if node == goal:  # Stop if goal is found
            print("\nGoal found!")
            return
        
        for neighbour in graph[node]:  # Visit each neighbour
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

# Define Start and Goal Nodes
start_node = 'A'
goal_node = 'M'

# Run BFS
print("\nFollowing is the Breadth-First Search (BFS):")
bfs(tree, start_node, goal_node)
```

```python
'''DFS Code Explaination'''
# Graph Representation
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G'],
    'D': ['H', 'I'],
    'E': ['J'],
    'F': ['K', 'L'],
    'G': [],
    'H': [],
    'I': ['M', 'N'],
    'J': [],
    'K': [],
    'L': [],
    'M': [],
    'N': [],
}

# DFS Function
def dfs(graph, start, goal):
    visited = []  # List to keep track of visited nodes to avoid revisiting
    stack = []    # Initialize a stack to help in the depth-first traversal
    
    stack.append(start)  # Push the start node onto the stack

    # Start DFS loop
    while stack:
        # Pop the top node from the stack (this is the current node we're exploring)
        node = stack.pop()
        
        # Check if this node has already been visited
        if node not in visited:
            visited.append(node)  # Mark this node as visited
            
            print(f"Visited {node}")  # Print the node when it's visited
            
            # If we've found the goal, stop the search
            if node == goal:
                print("\nGoal found!")
                return  # Exit the function once we find the goal

            # Add all unvisited neighbors to the stack (we will explore them next)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    stack.append(neighbor)  # Push unvisited neighbors onto the stack

# Define Start and Goal Nodes
start_node = 'A'
goal_node = 'M'

# Run DFS
print("\nFollowing is the Depth-First Search (DFS):")
dfs(graph, start_node, goal_node)
```

```python
'''Depth Limited Search!'''
'''Imagine we're in a magical garden (the graph), and we're on a treasure hunt (finding the goal).
But... here‚Äôs the twist: we can only explore the garden to a certain depth ‚Äî just like how you can
only dig a little bit before hitting a treasure chest in this specific spot.
So, let‚Äôs explore how Depth-Limited Search (DLS) works with the help of some cute code! üê±üíñ'''
# The garden is a graph where each plant (node) has its own branches (neighbors).
garden = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G'],
    'D': ['H', 'I'],
    'E': ['J'],
    'F': ['K', 'L'],
    'G': [],
    'H': [],
    'I': ['M', 'N'],
    'J': [],
    'K': [],
    'L': [],
    'M': [],
    'N': [],
}
'''Here, the garden has many plants, and each plant has its neighbors (branches).
 But, just like you have a "maximum depth" for digging in the garden, 
 Depth-Limited Search (DLS) only lets us dig (or explore) to a specific depth.'''
# Depth-Limited Search Function (DLS)
def dls(garden, start, goal, depth_limit):
    visited = []  # To track which plants we've already visited
    stack = []    # Our digging tools (a stack to explore)
    
    # Start at the starting plant (node)
    stack.append((start, 0))  # We store both plant and its depth
    visited.append(start)
    
    while stack:
        current_plant, current_depth = stack.pop()  # Pop the last plant to explore
        
        # Check if we've reached the goal (yay! treasure!)
        if current_plant == goal:
            print(f"\nYay! Found the treasure at plant {current_plant}! üéâ")
            return
        
        # If we're not at the goal yet and haven't reached the depth limit
        if current_depth < depth_limit:
            print(f"Exploring plant {current_plant} at depth {current_depth}... üåø")
            
            # Visit the neighbors (branches) of the current plant
            for neighbor in garden[current_plant]:
                if neighbor not in visited:
                    visited.append(neighbor)
                    stack.append((neighbor, current_depth + 1))  # Increase depth for neighbors

    print("\nOops, no treasure found within the depth limit! üò¢")

# Starting plant, goal plant, and depth limit
start_plant = 'A'
goal_plant = 'M'
depth_limit = 3  # Let's limit our digging to depth 3!

# Let's go treasure hunting in the garden with DLS! ‚ú®
print("\nStarting our treasure hunt in the garden using Depth-Limited Search (DLS)...")
dls(garden, start_plant, goal_plant, depth_limit)
```
